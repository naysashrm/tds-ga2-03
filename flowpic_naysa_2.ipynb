```python
import pylab as pl
import matplotlib.cm as cm
from mpl_toolkits.axes_grid1 import make_axes_locatable
import numpy.ma as ma
import numpy as np # Ensure numpy is imported
import random # Often useful for shuffling, etc.
import os # Useful for path operations

# --- Keras/TensorFlow Imports ---
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D, BatchNormalization, Activation
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import backend as K
from tensorflow.keras.metrics import top_k_categorical_accuracy

print("All necessary libraries imported.")


IMAGE_HEIGHT = 1500
IMAGE_WIDTH = 1500
NUM_CHANNELS = 1

# Define the input shape for your Keras model based on 'channels_last' (NHWC)
input_shape = (IMAGE_HEIGHT, IMAGE_WIDTH, NUM_CHANNELS)
print(f"Defined Model Input Shape (channels_last): {input_shape}")

# Define the number of classes (from your multi-class dataset)
num_classes = 5 # (Browse, chat, file_transfer, video, voip)
print(f"Defined Number of Classes: {num_classes}")

# Define a model name for saving plots and the model itself
MODEL_NAME = "flowpic_multiclass_cnn"
print(f"Defined Model Name: {MODEL_NAME}")

# Ensure Keras backend is set to 'channels_last'
if K.image_data_format() != 'channels_last':
    print("Keras image data format was not 'channels_last'. Setting it now.")
    K.set_image_data_format('channels_last')
print(f"Keras image data format: {K.image_data_format()}")

# --- Custom Metrics Definitions ---
def precision(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision

def recall(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + K.epsilon())
    return recall

def f1_score(y_true, y_pred):
    prec = precision(y_true, y_pred)
    rec = recall(y_true, y_pred)
    # Handle potential division by zero if prec+rec is 0
    if K.is_keras_tensor(prec) or K.is_keras_tensor(rec):
        return 2 * ((prec * rec) / (prec + rec + K.epsilon()))
    else: # Fallback for non-Keras tensors, though unlikely in this context
        return 2 * ((prec * rec) / (prec + rec + np.finfo(float).eps))

def top_2_categorical_accuracy(y_true, y_pred):
    return top_k_categorical_accuracy(y_true, y_pred, k=2)

print("Custom metrics (precision, recall, f1_score, top_2_categorical_accuracy) defined.")


# --- Model Definition ---
model = Sequential()
print("\n--- Building Model Architecture ---")

# First Convolutional Block
# input_shape is now (1500, 1500, 1)
model.add(Conv2D(10, kernel_size=(10, 10), strides=5, padding="same", input_shape=input_shape))
convout1 = Activation('relu')
model.add(convout1)
# Output for (1500,1500,1) with strides 5, padding same, 10 filters will be (300, 300, 10)
print(f"After Conv2D_1 and Activation_1: {model.output_shape}")

model.add(MaxPooling2D(pool_size=(2, 2)))
# Output for (300,300,10) with pool 2x2 will be (150, 150, 10)
print(f"After MaxPooling2D_1: {model.output_shape}")

# Second Convolutional Block
# Input for this layer is (150, 150, 10)
model.add(Conv2D(20, (10, 10), strides=5, padding="same"))
convout2 = Activation('relu') # Named for easy access in visualization
model.add(convout2)
# Output for (150,150,10) with strides 5, padding same, 20 filters will be (30, 30, 20)
print(f"After Conv2D_2 and Activation_2: {model.output_shape}")

model.add(Dropout(0.25))
print(f"After Dropout_1: {model.output_shape}") # Shape remains the same

model.add(MaxPooling2D(pool_size=(2, 2)))
# Output for (30,30,20) with pool 2x2 will be (15, 15, 20)
print(f"After MaxPooling2D_2: {model.output_shape}")

# Fully Connected Layers
model.add(Flatten())
# Output for (15, 15, 20) will be (15 * 15 * 20) = 4500
print(f"After Flatten: {model.output_shape}")

model.add(Dense(64, activation='relu'))
print(f"After Dense_1: {model.output_shape}") # Expected: (None, 64)

model.add(Dropout(0.5))
print(f"After Dropout_2: {model.output_shape}") # Shape remains the same

model.add(Dense(num_classes, activation='softmax')) # Output layer for multi-class classification
print(f"After Dense_2 (Output Layer): {model.output_shape}") # Expected: (None, 5)
print("--- Model Architecture Built Successfully ---")

# --- Model Compilation ---
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy', top_2_categorical_accuracy, f1_score, precision, recall])
print("Model compiled successfully with 'categorical_crossentropy' loss, 'adam' optimizer, and custom metrics.")


# --- Plotting Utility Functions ---
def nice_imshow(ax, data, vmin=None, vmax=None, cmap=None, bar=True):
    """Wrapper around pl.imshow"""
    if cmap is None:
        cmap = cm.jet
    if vmin is None:
        vmin = data.min()
    if vmax is None:
        vmax = data.max()
    divider = make_axes_locatable(ax)
    im = ax.imshow(data, vmin=vmin, vmax=vmax, interpolation='nearest', cmap=cmap, origin='lower')
    if bar:
        cax = divider.append_axes("right", size="5%", pad=0.05)
        pl.colorbar(im, cax=cax)

def plotNNFilter2(data, nrows, ncols, layer_name, cmap=None, bar=True):
    """Wrapper around pl.subplot with color bar"""
    if cmap is None:
        cmap = "gray"

    fig, axes = pl.subplots(nrows, ncols, figsize=(5*ncols, 4*nrows))
    axes = axes.flatten() # Flatten the axes array for easy iteration

    for i, ax in enumerate(axes):
        if i < data.shape[2]: # Ensure we don't go out of bounds if num_filters < nrows*ncols
            im = ax.imshow(data[:,:,i], interpolation="nearest", cmap=cmap)
            ax.set_yticklabels([])
            ax.set_xticklabels([])
            ax.invert_yaxis()
        else: # Hide empty subplots if there are more subplots than filters
            fig.delaxes(ax) # Remove axis from figure

    fig.subplots_adjust(wspace=0.025, hspace=0.05)
    if bar:
        # Create a single colorbar for all subplots
        cbar_ax = fig.add_axes([0.92, 0.15, 0.02, 0.7]) # [left, bottom, width, height]
        fig.colorbar(im, cax=cbar_ax)

    pl.savefig(MODEL_NAME +  "_plotNNFilter2_" + layer_name + ".png", bbox_inches='tight', pad_inches=1)
    pl.show()

def plotNNFilter(data, nrows, ncols, layer_name, cmap=None, bar=True):
    """Wrapper around pl.subplot"""
    if cmap is None:
        cmap = "gray"

    pl.figure(figsize=(3*ncols, 3*nrows))

    for i in range(nrows*ncols):
        if i < data.shape[2]: # Ensure we don't go out of bounds
            pl.subplot(nrows, ncols, i+1)
            pl.imshow(data[:,:,i], interpolation="nearest", cmap=cmap)
            pl.xticks([])
            pl.yticks([])
            pl.gca().invert_yaxis()
    pl.subplots_adjust(wspace=0.025, hspace=0.05)
    pl.savefig(MODEL_NAME +  "_plotNNFilter_" + layer_name + ".png", bbox_inches='tight', pad_inches=1)
    pl.show()

def make_mosaic(imgs, nrows, ncols, border=1):
    """
    Given a set of images with all the same shape, makes a
    mosaic with nrows and ncols
    """
    nimgs = imgs.shape[2] # Number of images (filters)
    imshape = imgs.shape[0:2] # Height and width of each image

    mosaic = ma.masked_all((nrows * imshape[0] + (nrows - 1) * border,
                            ncols * imshape[1] + (ncols - 1) * border),
                            dtype=np.float32)

    paddedh = imshape[0] + border
    paddedw = imshape[1] + border
    for i in range(nimgs):
        row = int(np.floor(i / ncols))
        col = i % ncols

        mosaic[row * paddedh:row * paddedh + imshape[0],
               col * paddedw:col * paddedw + imshape[1]] = imgs[:,:,i]
    return mosaic

def mosaic_imshow(imgs, nrows, ncols, cmap=None, border=1, layer_name="convout"):
    pl.figure(figsize=(3*ncols, 3*nrows))
    nice_imshow(pl.gca(), make_mosaic(imgs, nrows, ncols, border=border), cmap=cmap)
    pl.savefig(MODEL_NAME +  "_mosaic_imshow_" + layer_name + ".png", bbox_inches='tight', pad_inches=1)
    pl.show()

print("Plotting utility functions defined.")


# ---get_layer_output function ---
def get_layer_output(layer, input_img, layer_name):
    # Create a sub-model that takes the original model's input
    # and outputs the desired intermediate layer's output.
    layer_model = Model(inputs=model.inputs, outputs=layer.output)

    # Get the output by predicting on the input image
    # input_img should already be (batch, H, W, C) for channels_last
    C = layer_model.predict(input_img)

    # After prediction, C will have shape (batch_size, H, W, num_filters) for channels_last
    # Squeeze removes the batch dimension if it's 1.
    # Resulting shape will be (H, W, num_filters) which is suitable for plotting
    C = np.squeeze(C)
    print(f"Inside get_layer_output for {layer_name}: C.shape after squeeze: {C.shape}")

    # No transpose is needed here as the output is already in (Height, Width, Channels) format
    # for plotting functions expecting (H, W, Filters)
    return C

print("Corrected get_layer_output function defined using tensorflow.keras.models.Model and channels_last output.")

```

```python
import tensorflow as tf
import pylab as pl
import matplotlib.cm as cm
from mpl_toolkits.axes_grid1 import make_axes_locatable
import numpy.ma as ma
import numpy as np
import random
import os

# --- Keras/TensorFlow Imports ---
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D, BatchNormalization, Activation
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import backend as K
from tensorflow.keras.metrics import top_k_categorical_accuracy
from tensorflow.keras.callbacks import TensorBoard, ModelCheckpoint

print("All necessary libraries imported.")


# --- Essential Definitions (Confirmed from your data inspection) ---
IMAGE_HEIGHT = 1500
IMAGE_WIDTH = 1500
NUM_CHANNELS = 1 # For grayscale FlowPics

num_classes = 5 # (browsing, chat, file_transfer, video, voip) - Confirmed as 5

MODEL_NAME = "flowpic_multiclass_cnn"
print(f"Defined Model Input Shape (channels_last): ({IMAGE_HEIGHT}, {IMAGE_WIDTH}, {NUM_CHANNELS})")
print(f"Defined Number of Classes: {num_classes}")
print(f"Defined Model Name: {MODEL_NAME}")

# Ensure Keras backend is set to 'channels_last'
if K.image_data_format() != 'channels_last':
    print("Keras image data format was not 'channels_last'. Setting it now.")
    K.set_image_data_format('channels_last')
print(f"Keras image data format: {K.image_data_format()}")


# --- Custom Metrics Definitions ---
def precision(y_true, y_pred):
    y_true = K.cast(y_true, K.floatx())
    y_pred = K.cast(y_pred, K.floatx())
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision

def recall(y_true, y_pred):
    y_true = K.cast(y_true, K.floatx())
    y_pred = K.cast(y_pred, K.floatx())
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + K.epsilon())
    return recall

def f1_score(y_true, y_pred):
    prec = precision(y_true, y_pred)
    rec = recall(y_true, y_pred)
    if K.is_keras_tensor(prec) or K.is_keras_tensor(rec):
        return 2 * ((prec * rec) / (prec + rec + K.epsilon()))
    else:
        return 2 * ((prec * rec) / (prec + rec + np.finfo(float).eps))

def top_2_categorical_accuracy(y_true, y_pred):
    y_true = K.cast(y_true, K.floatx())
    y_pred = K.cast(y_pred, K.floatx())
    return top_k_categorical_accuracy(y_true, y_pred, k=2)

print("Custom metrics (precision, recall, f1_score, top_2_categorical_accuracy) defined.")


# --- Model Definition ---
# Input shape now uses the confirmed values
input_shape = (IMAGE_HEIGHT, IMAGE_WIDTH, NUM_CHANNELS)

model = Sequential()
print("\n--- Building Model Architecture ---")
model.add(Conv2D(10, kernel_size=(10, 10), strides=5, padding="same", input_shape=input_shape))
convout1 = Activation('relu')
model.add(convout1)
print(f"After Conv2D_1 and Activation_1: {model.output_shape}")

model.add(MaxPooling2D(pool_size=(2, 2)))
print(f"After MaxPooling2D_1: {model.output_shape}")

model.add(Conv2D(20, (10, 10), strides=5, padding="same"))
convout2 = Activation('relu')
model.add(convout2)
print(f"After Conv2D_2 and Activation_2: {model.output_shape}")

model.add(Dropout(0.25))
print(f"After Dropout_1: {model.output_shape}")

model.add(MaxPooling2D(pool_size=(2, 2)))
print(f"After MaxPooling2D_2: {model.output_shape}")

model.add(Flatten())
print(f"After Flatten: {model.output_shape}")

model.add(Dense(64, activation='relu'))
print(f"After Dense_1: {model.output_shape}")

model.add(Dropout(0.5))
print(f"After Dropout_2: {model.output_shape}")

model.add(Dense(num_classes, activation='softmax')) # num_classes confirmed as 5
print(f"After Dense_2 (Output Layer): {model.output_shape}")
print("--- Model Architecture Built Successfully ---")

# --- Model Compilation ---
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy', top_2_categorical_accuracy, f1_score, precision, recall])
print("Model compiled successfully with 'categorical_crossentropy' loss, 'adam' optimizer, and custom metrics.")


# --- Plotting Utility Functions (Same as before) ---
def nice_imshow(ax, data, vmin=None, vmax=None, cmap=None, bar=True):
    if cmap is None:
        cmap = cm.jet
    if vmin is None:
        vmin = data.min()
    if vmax is None:
        vmax = data.max()
    divider = make_axes_locatable(ax)
    im = ax.imshow(data, vmin=vmin, vmax=vmax, interpolation='nearest', cmap=cmap, origin='lower')
    if bar:
        cax = divider.append_axes("right", size="5%", pad=0.05)
        pl.colorbar(im, cax=cax)

def plotNNFilter2(data, nrows, ncols, layer_name, cmap=None, bar=True):
    if cmap is None:
        cmap = "gray"
    fig, axes = pl.subplots(nrows, ncols, figsize=(5*ncols, 4*nrows))
    axes = axes.flatten()
    for i, ax in enumerate(axes):
        if i < data.shape[2]:
            im = ax.imshow(data[:,:,i], interpolation="nearest", cmap=cmap)
            ax.set_yticklabels([])
            ax.set_xticklabels([])
            ax.invert_yaxis()
        else:
            fig.delaxes(ax)
    fig.subplots_adjust(wspace=0.025, hspace=0.05)
    if bar:
        cbar_ax = fig.add_axes([0.92, 0.15, 0.02, 0.7])
        fig.colorbar(im, cax=cbar_ax)
    pl.savefig(MODEL_NAME +  "_plotNNFilter2_" + layer_name + ".png", bbox_inches='tight', pad_inches=1)
    pl.show()

def plotNNFilter(data, nrows, ncols, layer_name, cmap=None, bar=True):
    if cmap is None:
        cmap = "gray"
    pl.figure(figsize=(3*ncols, 3*nrows))
    for i in range(nrows*ncols):
        if i < data.shape[2]:
            pl.subplot(nrows, ncols, i+1)
            pl.imshow(data[:,:,i], interpolation="nearest", cmap=cmap)
            pl.xticks([])
            pl.yticks([])
            pl.gca().invert_yaxis()
    pl.subplots_adjust(wspace=0.025, hspace=0.05)
    pl.savefig(MODEL_NAME +  "_plotNNFilter_" + layer_name + ".png", bbox_inches='tight', pad_inches=1)
    pl.show()

def make_mosaic(imgs, nrows, ncols, border=1):
    nimgs = imgs.shape[2]
    imshape = imgs.shape[0:2]
    mosaic = ma.masked_all((nrows * imshape[0] + (nrows - 1) * border,
                            ncols * imshape[1] + (ncols - 1) * border),
                            dtype=np.float32)
    paddedh = imshape[0] + border
    paddedw = imshape[1] + border
    for i in range(nimgs):
        row = int(np.floor(i / ncols))
        col = i % ncols
        mosaic[row * paddedh:row * paddedh + imshape[0],
               col * paddedw:col * paddedw + imshape[1]] = imgs[:,:,i]
    return mosaic

def mosaic_imshow(imgs, nrows, ncols, cmap=None, border=1, layer_name="convout"):
    pl.figure(figsize=(3*ncols, 3*nrows))
    nice_imshow(pl.gca(), make_mosaic(imgs, nrows, ncols, border=border), cmap=cmap)
    pl.savefig(MODEL_NAME +  "_mosaic_imshow_" + layer_name + ".png", bbox_inches='tight', pad_inches=1)
    pl.show()

print("Plotting utility functions defined.")


# --- Corrected get_layer_output function ---
def get_layer_output(layer, input_img, layer_name):
    layer_model = Model(inputs=model.inputs, outputs=layer.output)
    C = layer_model.predict(input_img)
    C = np.squeeze(C)
    print(f"Inside get_layer_output for {layer_name}: C.shape after squeeze: {C.shape}")
    return C

print("Corrected get_layer_output function defined using tensorflow.keras.models.Model and channels_last output.")


# --- Visualization Code ---
print("\n--- Starting Visualization ---")
i = 35
X_display = x_train[i][0]

X_input_for_model = np.transpose(x_train[i:i+1], (0, 2, 3, 1))

print(f"Selected input image index: {i}")
print(f"Selected input image for display (X_display) shape: {X_display.shape}")
print(f"Selected input image for model (X_input_for_model) shape: {X_input_for_model.shape}")
print(f"True label for selected image: {y_train_true[i]}")

pl.figure(figsize=(15, 15))
pl.title(f'Input Image (Index: {i}, True Label: {y_train_true[i]})')
nice_imshow(pl.gca(), X_display, vmin=0, vmax=1, cmap=cm.binary)
pl.savefig(MODEL_NAME +  "_input_image_idx_" + str(i) + ".png", bbox_inches='tight', pad_inches=1)
pl.show()
print(f"Input image {i} saved as {MODEL_NAME}_input_image_idx_{i}.png")

print("\n--- Visualizing convout1 output ---")
C1 = get_layer_output(convout1, X_input_for_model, layer_name="convout1_output")
mosaic_imshow(C1, 2, 5, cmap=cm.binary, border=2, layer_name="convout1_output_mosaic")
plotNNFilter(C1, 2, 5, cmap=cm.binary, layer_name="convout1_output_filter")
plotNNFilter2(C1, 2, 5, cmap=cm.binary, layer_name="convout1_output_filter2")
print("convout1 visualizations generated and saved.")

print("\n--- Visualizing convout2 output ---")
C2 = get_layer_output(convout2, X_input_for_model, layer_name="convout2_output")
mosaic_imshow(C2, 4, 5, cmap=cm.binary, border=2, layer_name="convout2_output_mosaic")
plotNNFilter(C2, 4, 5, cmap=cm.binary, layer_name="convout2_output_filter")
plotNNFilter2(C2, 4, 5, cmap=cm.binary, layer_name="convout2_output_filter2")
print("convout2 visualizations generated and saved.")

print("\n--- All visualizations completed. ---")


# --- Training Configuration and Execution ---
batch_size = 32
epochs = 10

if y_train.dtype != np.float32:
    print(f"Converting y_train from {y_train.dtype} to float32...")
    y_train = y_train.astype(np.float32)
if y_val.dtype != np.float32:
    print(f"Converting y_val from {y_val.dtype} to float32...")
    y_val = y_val.astype(np.float32)
# Also ensure x_train and x_val are float32, as common for image data
if x_train.dtype != np.float32:
    print(f"Converting x_train from {x_train.dtype} to float32...")
    x_train = x_train.astype(np.float32)
if x_val.dtype != np.float32:
    print(f"Converting x_val from {x_val.dtype} to float32...")
    x_val = x_val.astype(np.float32)

# Verify data shapes and types after conversion
print(f"x_train shape after conversion: {x_train.shape}, dtype: {x_train.dtype}")
print(f"y_train shape after conversion: {y_train.shape}, dtype: {y_train.dtype}")
print(f"x_val shape after conversion: {x_val.shape}, dtype: {x_val.dtype}")
print(f"y_val shape after conversion: {y_val.shape}, dtype: {y_val.dtype}")


# Callbacks for TensorBoard and Model Checkpointing
tensorboard = TensorBoard(log_dir='./Graph', histogram_freq=1, write_graph=True,
                          write_images=True)
checkpointer_loss = ModelCheckpoint(filepath= MODEL_NAME + '_loss.weights.h5', verbose=1, save_best_only=True, save_weights_only=True)
checkpointer_acc = ModelCheckpoint(monitor='val_accuracy', filepath= MODEL_NAME + '_acc.weights.h5', verbose=1, save_best_only=True, save_weights_only=True)


# Define the generator function
def generator(features, labels, batch_size):
    num_samples = len(features)
    index = 0
    while True:
        if index + batch_size > num_samples:
            batch_features = np.append(features[index:num_samples], features[0:(index + batch_size) - num_samples], axis=0)
            batch_labels = np.append(labels[index:num_samples], labels[0:(index + batch_size) - num_samples], axis=0)
            index = (index + batch_size) - num_samples
        else:
            batch_features = features[index:index + batch_size]
            batch_labels = labels[index:index + batch_size]
            index += batch_size

        batch_features_channels_last = np.transpose(batch_features, (0, 2, 3, 1))

        yield batch_features_channels_last, batch_labels

# Calculate steps_per_epoch for model.fit
steps_per_epoch = len(x_train) // batch_size


# --- Create tf.data.Dataset from your generator ---
# Print values used in output_signature for debugging
print(f"Values for output_signature: IMAGE_HEIGHT={IMAGE_HEIGHT}, IMAGE_WIDTH={IMAGE_WIDTH}, NUM_CHANNELS={NUM_CHANNELS}, num_classes={num_classes}")

# Define the output signature (data types and shapes) of your generator
output_signature = (
    tf.TensorSpec(shape=(None, IMAGE_HEIGHT, IMAGE_WIDTH, NUM_CHANNELS), dtype=tf.float32),
    tf.TensorSpec(shape=(None, num_classes), dtype=tf.float32)
)

dataset = tf.data.Dataset.from_generator(
    lambda: generator(x_train, y_train, batch_size),
    output_signature=output_signature
)

dataset = dataset.prefetch(tf.data.AUTOTUNE)


# Train the model using model.fit with the tf.data.Dataset
print("\n--- Starting Model Training ---")
history = model.fit(
    dataset,
    epochs=epochs,
    steps_per_epoch=steps_per_epoch,
    verbose=1,
    callbacks=[tensorboard, checkpointer_loss, checkpointer_acc],
    validation_data=(np.transpose(x_val, (0, 2, 3, 1)).astype(np.float32), y_val)
)

print("\n--- Model Training Completed ---")
```

```python
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
import itertools
from tensorflow.keras.utils import to_categorical # Import for one-hot encoding
try:
    y_val_true
except NameError:
    y_val_true = np.array([0]*61 + [1]*70 + [2]*63 + [3]*54 + [4]*66)
    np.random.shuffle(y_val_true)
    print("Using dummy y_val_true for demonstration.")

try:
    y_val_prediction
except NameError:
    # Dummy predictions for demonstration (will have some errors to show confusion)
    y_val_prediction = np.copy(y_val_true)
    # Introduce some misclassifications for testing the stats function
    y_val_prediction[50:55] = 3
    y_val_prediction[100:103] = 4
    print("Using dummy y_val_prediction for demonstration.")

try:
    num_classes
except NameError:
    num_classes = len(np.unique(y_val_true)) # Infer num_classes from true labels
    print(f"Inferred num_classes: {num_classes}")

def stats(y_true_one_hot, y_pred_integer):
    """
    Calculates and prints overall accuracy and per-class accuracy.
    Assumes y_true_one_hot is one-hot encoded and y_pred_integer contains integer class predictions.
    """
    # Overall accuracy
    correct_overall = sum([1 for i, pred in enumerate(y_pred_integer) if y_true_one_hot[i][pred] == 1])
    total_overall = y_true_one_hot.shape[0]
    overall_accuracy = correct_overall * 1.0 / total_overall if total_overall > 0 else 0.0
    print(f"Total samples: {total_overall}, Correctly predicted overall: {correct_overall}, Overall Accuracy: {overall_accuracy:.4f}")

    # Per-class accuracy
    for class_ind in range(y_true_one_hot.shape[1]): # Iterate through each class index
        total_ind = len([1 for val in y_true_one_hot if val[class_ind] == 1])
        correct_ind = sum([1 for i, pred in enumerate(y_pred_integer) if (pred == class_ind and y_true_one_hot[i][pred] == 1)])

        class_accuracy = correct_ind * 1.0 / total_ind if total_ind > 0 else 0.0
        print(f"Class {class_ind}: Total {total_ind}, Correct {correct_ind}, Accuracy: {class_accuracy:.4f}")


# Convert integer labels (y_val_true) to one-hot encoded format (y_val_one_hot)
y_val_one_hot = to_categorical(y_val_true, num_classes).astype(np.float32)

print("\n--- Running custom `stats` function ---")
stats(y_val_one_hot, y_val_prediction)

print("\n--- Running `classification_report` (for comparison) ---")
print(classification_report(y_val_true, y_val_prediction))
```
